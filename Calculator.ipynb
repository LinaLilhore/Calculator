```python
a = int (input("Enter the first value:" ))
b = int (input("Enter the second value:" ))  
n=int(input("what do you want to perform , enter-1(For addition), 2(For sutraction), 3(For Modulas),  4(For multiplication),5(For divis):"))
if(n==1):
    ans= a+b
elif(n==2):
   ans= a-b  
elif(n==3):
    ans= a%b
elif(n==4):
    ans= a*b
elif(n==5):
   ans= a/b  
print("Your answer = " , ans)
```
